# **SPLIITING THE FEATURES AND TARGET**

**in x we storing all features except target and in target we storing target**

```python
X=heart_data.drop(columns='target',axis=1)
Y=heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

# **SPLITTING TGE DATA INTO TRAIN AND TEST DATA**

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# **MODEL TRAINING USING LOGISTIC REGRESSION MODEL**

```python
model=LogisticRegression()
```

#training the logisticRegression model with training data

```python
model.fit(X_train,Y_train)
```

# **model evaluation**

**ACCURACY SCORE**

```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
#we asking model to predict target values and we will store in X_train_prediction
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
#it will compare X_train_prediction with Y_train values
```

```python
print('ACCURACY ON TRAINING DATA',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
#we asking model to predict target values and we will store in X_test_prediction
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
#it will compare X_test_prediction with Y_test values
```

```python
print('ACCURACY ON TEST DATA',test_data_accuracy)
```
